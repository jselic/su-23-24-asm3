```python
import pandas as pd
import re
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, Ridge, Lasso, RidgeCV, LassoCV
from sklearn.model_selection import KFold, cross_val_score, cross_val_predict, train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import MinMaxScaler
from scipy.stats import probplot
import numpy as np
```

```python
df = pd.read_csv("data/communities.data")
```

```python
with open("data/communities.names") as f:
    text = f.readlines()
    attrs = [el.split(' ')[1] for el in text if re.match(r'@attribute',el)]
    
df.columns = attrs

#removing non-predictive columns
df = df.drop(['state', 'county', 'community', 'communityname', 'fold'],axis=1)
df = df.applymap(lambda x: x if x=='?' else float(x))
specific_columns = df.apply(lambda x: x.astype(str).str.contains('\?', na=False).any())
specific_columns= list(specific_columns[specific_columns==True].index)
df['specializedParameters'] = df['PolicCars'].apply(lambda x: 1 if x!='?' else 0)
df[specific_columns] = df[specific_columns].applymap(lambda x: x if x!='?' else -1);
```

```python
def best_attributes(X,y,model,n_attributes=12,metric=mean_squared_error):
    selected_attributes = []
    attrs = list(X.columns)
    k = KFold()
    
    while len(selected_attributes) < n_attributes and attrs:
        scores = dict(
            (att, np.mean(cross_val_score(model,X[selected_attributes + [att]],y,cv=k)))
        for att in attrs)
        best = max(scores, key=scores.get)
        selected_attributes.append(best)
        attrs.remove(best)
        
    return selected_attributes
```

## Comparisons - communities and crime

```python
linear_model = LinearRegression()
X = df.drop('ViolentCrimesPerPop',axis=1)
y = df['ViolentCrimesPerPop']
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
best_12 = best_attributes(X_train, y_train,linear_model)
df_reduced = df[best_12 + ['ViolentCrimesPerPop']]
```

```python
X = df_reduced.drop('ViolentCrimesPerPop',axis=1)
y = df_reduced['ViolentCrimesPerPop']
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
best_12
```

### Linear regression

```python
linear_model= LinearRegression()
linear_model.fit(X_train, y_train)
y_pred = linear_model.predict(X_test)
print(f"MSE: {mean_squared_error(y_test,y_pred)}")
linear_residuals = y_test - y_pred
```

```python
probplot(linear_residuals, plot=sns.mpl.pyplot)
plt.title('Normal Q-Q Plot')
plt.show()
```

### Ridge Regression

```python
X = df.drop('ViolentCrimesPerPop',axis=1)
y = df['ViolentCrimesPerPop']
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
parameters = [10**i for i in range(-1,7)] 
ridge_model = RidgeCV(alphas=parameters)
ridge_model.fit(X_train, y_train)
y_pred = ridge_model.predict(X_test)
print(f"MSE: {mean_squared_error(y_test,y_pred)}")
ridge_residuals = y_test - y_pred
print(f"Optimal reg. parameter: {ridge_model.alpha_}")
```

```python
probplot(ridge_residuals, plot=sns.mpl.pyplot)
plt.title('Normal Q-Q Plot')
plt.show()
```

### Lasso Regression

```python
lasso_model = LassoCV(alphas=parameters)
lasso_model.fit(X_train, y_train)
y_pred = lasso_model.predict(X_test)
print(f"MSE: {mean_squared_error(y_test,y_pred)}")
lasso_residuals = y_test - y_pred
print(f"Optimal reg. parameter: {lasso_model.alpha_}")
```

```python
probplot(lasso_residuals, plot=sns.mpl.pyplot)
plt.title('Normal Q-Q Plot')
plt.show()
```

```python
X = df.drop('quality',axis=1)
y = df['quality']
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
class CustomRidgeRegression:
    def __init__(self, learning_rate=0.01, iterations=1000, alpha=1):
        self.learning_rate = learning_rate
        self.iterations = iterations
        self.alpha = alpha
    
    def fit(self, X,y):
        X = np.concatenate((np.ones((X.shape[0], 1)), X), axis=1)
        samples, features = X.shape
        self.weights = np.random.rand(features)
        
        for _ in range(self.iterations):
            y_pred = X @ self.weights
            gradient = (1/samples) * (X.T @ (y_pred - y)) + self.alpha * self.weights
            self.weights -=self.learning_rate * gradient 
            
    def predict(self, X):
        X = np.concatenate((np.ones((X.shape[0], 1)), X), axis=1)
        return X @ self.weights
 
    def get_params(self, deep=True):
        return {}
    
class SGDRidgeRegression(CustomRidgeRegression):        
    def fit(self, X,y):
        X = np.concatenate((np.ones((X.shape[0], 1)), X), axis=1)
        samples, features = X.shape
        self.weights = np.random.rand(features)
        
        for _ in range (self.iterations):
            ridx = np.random.randint(samples)
            X_i = X[ridx, :].reshape(1,-1)
            y_i = y.values[ridx]

            y_pred = X_i @ self.weights
            gradient = (1/samples) * (X_i.T @ (y_pred - y_i)) + self.alpha * self.weights
            self.weights -=self.learning_rate * gradient 
```

```python
model = CustomRidgeRegression(learning_rate=0.5,alpha=0.001)
# model = LinearRegression()
# model = SGDRidgeRegression(learning_rate=0.5,alpha=0.001)
```

```python
my_ridge_model = model
my_ridge_model.fit(X_train, y_train)
y_pred = my_ridge_model.predict(X_test)
print(f"MSE: {mean_squared_error(y_test,y_pred)}")
```
